```python
import pandas as pd
from sqlalchemy import create_engine, text
pd.options.display.max_rows = 100

ruta="C:/Users/Dell/OneDrive/Documentos/GitHub/Ent_DS/Ent_5/"
engine=create_engine('sqlite://')
print(engine)
df = pd.read_csv('DataScience_salaries_2024.csv').to_sql('Salaries_2024', engine, if_exists='replace', index=False)

def runQuery(sql):    
    result = engine.connect().execute((text(sql)))
    return pd.DataFrame(result.fetchall(), columns=result.keys())
```

```python
query1 = """SELECT experience_level, job_title ,company_size
FROM salaries_2024
WHERE company_size != 'M'"""
runQuery(query1)
```

```python
query2 = """SELECT experience_level, job_title AS ex_job
FROM salaries_2024
WHERE company_size != 'M'
ORDER BY ex_job ASC
"""
runQuery(query2)
```

```python
query3 = """SELECT DISTINCT job_title
FROM salaries_2024
WHERE job_title LIKE '%engineer%'
ORDER BY job_title DESC"""
runQuery(query3)
```

```python
query4 = """SELECT job_title, experience_level, salary_in_usd AS Salario_Dolares,
    CASE
        WHEN salary_in_usd >= 21013 THEN 'Yes'
        WHEN salary_in_usd <  21013 THEN 'No'
        ELSE 'Missing Data'
    END AS Salarie_Validar
FROM salaries_2024
ORDER BY experience_level DESC"""
runQuery(query4)
```
